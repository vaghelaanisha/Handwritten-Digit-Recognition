```python
from mnist import MNIST
import numpy as np
from matplotlib import pyplot as plt
from operator import itemgetter, attrgetter
mndata = MNIST('D:\sjsu\ML_coursera\usc_ML\samples')

images, labels = mndata.load_testing()
images1, labels1 = mndata.load_training()

fig=plt.figure(figsize=(8,8))
rows=10
columns=6
count = 1

for i in range(0,10):
    first_image = images[i]
    first_image = np.array(first_image, dtype='float')
    dista =[[0,0,0]]
    
    for j in range(0,10000):
        second_image = images1[j]
        dist = np.sqrt(np.sum((second_image-first_image)**2))
        dista.append([dist,labels1[j],j])
        
    dista = np.array(dista)
    dista = dista[dista[:, 0].argsort()]
    pixels = first_image.reshape((28, 28))
    fig.add_subplot(rows, columns, count)
    count =count + 1
    plt.imshow(pixels, cmap='gray')
    
    for k in range(1,6):
        dista = dista.astype(np.int64)
        d = dista[k,2]
        second_image = images1[d]
        second_image = np.array(second_image)
        pixels = second_image.reshape((28, 28))
        fig.add_subplot(rows, columns, count)
        count = count + 1
        plt.imshow(pixels, cmap='gray')
        
plt.show()
```
