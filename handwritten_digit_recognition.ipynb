```python
from mnist import MNIST
import numpy as np
from matplotlib import pyplot as plt
import random as rand
mndata = MNIST('.')

images, labels = mndata.load_testing()
images1, labels1 = mndata.load_training()

fig=plt.figure(figsize=(8,8))
rows=5
columns=5
for i in range(1,26):
    index = rand.randrange(0, len(images))  
    first_image = images[index]
    first_image = np.array(first_image, dtype='float')
    pixels = first_image.reshape((28, 28))
    fig.add_subplot(rows, columns, i)
    plt.imshow(pixels, cmap='gray')
plt.show()
```

```python
from sklearn.neighbors import KNeighborsClassifier
neigh = KNeighborsClassifier(n_neighbors=5)
neigh.fit(images1, labels1) 
index = []
index = []
for i in range(0,10):
    # print(neigh.predict([images[i]]))
    dist, ind = neigh.kneighbors([images[i]], 5)
    ind = np.array(ind)
    ind=ind.ravel()
    index.append(ind)
    
# print(index)
    
fig1=plt.figure(figsize=(8,8))
rows=10
columns=5
count=1
for j in range(0,10):
    for k in range(0,5):
        first_image = images1[index[j][k]]
        first_image = np.array(first_image, dtype='float')
        pixels = first_image.reshape((28, 28))
        fig1.add_subplot(rows, columns, count)
        count = count + 1
        plt.imshow(pixels, cmap='gray')
plt.show()
    
```
